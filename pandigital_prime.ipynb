```python
from math import sqrt
```

```python
def e_numero_primo(n: int)->bool:
    """Função que define se o número n é primo.  
    
    """
    d = 3
    c = 0
    while d <= int(sqrt(n))+1:
        c += 1 if (n % d == 0) else 0
        d += 1
        if c >=1: return False
    
    return c == 0
```

- O desafio quer o maior número pandigital, que tenha todos os números ao menos uma vez, até o número de dígitos.  
- Ou seja, se tem 9 dígitos, tem que ter os 9 números (1 a 9). 
- Se tem 8 dígitos, tem que ter os números 1 a 8, e assim sucessivamente. 
- Um detalhe é que se pensarmos nos 9 dígitos (9,8,7,6,5,4,3,2,1) então qualquer combinação desses números a soma desses números vai ser (9+8+7+6+5+4+3+2+1 = 45 -> 4+5 = 9) ou seja, qualquer combinação desses números vai ser um divisível por 3. Já conseguimos ver com certeza que não existem números primos (pandigitais = somente um número em cada dígito) para nove dígitos.  
- O mesmo vale para 8 dígitos. (9+8+7+6+5+4+3+2+1 = 36 -> 3+6 = 9) também vai ser divisível por 3 qualquer combinação, logo não vai ter nenhum primo pandigital.
- Nos resta então fazer uma busca bruta de 7 dígitos para trás, pois o desafio quer o maior número. 
- Assim, podemos começar por: 7_654_321 até achar o primeiro primo. 

```python
def is_pandigital(n: int)->bool:
    l = len(str(n))

    gabarito = ''.join([str(v) for v in list(range(1, l+1))])
    numeros_ordenados = ''.join(sorted([v for v in str(n)]))

    return gabarito == numeros_ordenados

print (is_pandigital(2341))
print (is_pandigital(615234))
print (is_pandigital(123))
```

```python
achou_primo = False
n = 7_654_321 + 1
while not(achou_primo):
    n -= 1
    if is_pandigital(n):
        achou_primo = e_numero_primo(n)
    
print (n)
```
