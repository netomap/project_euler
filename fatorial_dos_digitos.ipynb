```python
from math import factorial
```

```python
def fatorial_dos_digitos(n: int):
    digitos = list(str(n))
    soma = 0
    for d in digitos:
        soma += factorial(int(d))
    
    return soma == n

numeros = []
n = 3
while (n <= 100_000_000):
    if (fatorial_dos_digitos(n)):
        numeros.append(n)
    
    if (n % 1_000_000 == 0):
        print (f'{n:,}, {numeros=}')
    
    n+=1
```

```python
sum(numeros)
```

- Por força bruta, fazendo uma simulação até 100_000_000, vou deduzir que só existem dois números que são esses da lista acima
