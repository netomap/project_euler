```python
MAX_NUMBER = 4_000_000

numeros = [1, 2]
n = 0

while n < MAX_NUMBER:
    n = sum(numeros[-2:])
    if (n < MAX_NUMBER):
        numeros.append(n)

print (f'Total de nmeros: {len(numeros)} at {MAX_NUMBER}')
```

```python
# Pegando todos os mpares
soma = 0
for n in numeros:
    if n % 2 == 0:
        soma += n

print (f'Soma = {soma} dos nmeros pares da sequncia de Fibonacci at {MAX_NUMBER}')
```
