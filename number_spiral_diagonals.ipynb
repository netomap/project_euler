- Para uma matriz 1x1, s tem o nmero 1
- Para uma matriz 3x3, as diagonais so 3, 5, 7, 9 (variam de 2 em 2) a partir do ltimo nmero da matriz anterior (que  1)
- Para uma matriz 5x5, as diagonais so: 13, 17, 21, 25 (variam de 4 em 4) a partir de 9, a ltima matriz (3x3)
- Para uma matriz 7x7, as diagonais so: 31, 37, 43, 49 (variam de 6 em 6) a partir de 25, matriz (5x5)

- Regra:
    1) Comeamos por dm (dimenso da matriz em 3) at 1001, variando de 2 em 2. Ex: 3, 5, 7, 9 .... 1001
    2) Definimos v (variao) comeando em 2, variando tambm de 2 em 2: Ex: 2, 4, 6, 8, ....
        - Repare que aqui v  igual a dm-1.
    3) Somamos 4 nmeros para cada iterao. Onde o primeiro nmero pn = v+(dm-1)^2
    

```python
ultima_dimensao = 1
soma_total = 0

for dm in range(3, 1001+1, 2):
    # print (f'{dm=}')
    # print (f'{ultima_dimensao=}')

    v = dm - 1
    # print (f'{v=}')

    pn = (ultima_dimensao*ultima_dimensao)
    # print (f'{pn=}')
    
    numeros = []
    for k in range(1, 5):
        numeros.append(pn + k*v)
        soma_total += pn+k*v

    # print (f'{numeros=}')

    ultima_dimensao = dm

    # print (30*'-')

soma_total += 1
# porque no foi considerado a matriz 1x1

print (f'{soma_total=}')
```
