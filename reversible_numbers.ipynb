- Procurar todos os nmeros que somados ao seu reverso possuem somente nmeros mpares no resultado.
- No podem zeros  esquerda.
- Existem 120 nmeros com essa caracterstica at 1000.
- Quantos existem at 10^9?

```python
import numpy as np
from tqdm import tqdm
from concurrent.futures import ThreadPoolExecutor
import multiprocessing
```

```python
def reverter_numero(n):
    reverso = 0
    while (n > 0):
        reverso = reverso*10 + (n % 10)
        n = n // 10
    
    return reverso

def verificar_se_tem_somente_numero_impar(n):
    while n > 0:
        if (n % 10) % 2 == 0: return False
        n = n // 10
    return True

def validar(n):
    if n % 10 == 0:
        return False
    
    return verificar_se_tem_somente_numero_impar(n + reverter_numero(n))

def contar_reversiveis(tupla):
    c = 0
    for n in range(tupla[0], tupla[1]):
        c += 1 if validar(n) else 0

    return c

print (contar_reversiveis((0, 1000)))
```

```python
c = 0
for n in tqdm(range(0, 1_000_000_000), ncols=100):
    c += 1 if validar(n) else 0

print (c)
```
