```python
from random import randint
import numpy as np
```

```python
# O maior nmero de trs dgitos  999, portanto o maior produto de nmeros de trs dgitos 
print (999 * 999)

# esse  o maior nmero possvel. Ou seja, o palndromo no pode ser maior que esse, tem que ser menor.
```

```python
def reverter_numero(n):
    numero_revertido = 0
    n, ultimo_numero =  n // 10, n % 10
    while n > 0:
        numero_revertido = numero_revertido*10 + ultimo_numero
        
        n, ultimo_numero =  n // 10, n % 10
    
    numero_revertido = numero_revertido*10 + ultimo_numero

    
    return numero_revertido

for k in range(10):
    n = randint(1_000, 1_000_000)
    print (n, reverter_numero(n), sep=' ==>> ')
```

```python
def numero_palindromo(n):
    n_rev = reverter_numero(n)
    return n == n_rev

for k in [987, 989, 1001, 8008, 45654, 4321234, 43211234, 5486]:
    print (k, numero_palindromo(k), sep=' == > ')
```

```python
print (999*999, '  o maior nmero produto de dois nmeros de trs dgitos')
print (997799, '  o maior palndrom possvel menor que o maior nmero do produto de trs dgitos')
print ('vamos verificar se  possvel encontrar um produto...')

d, n = 999, 997799
while n % d != 0:
    d -= 1

print (f'encontrado um nmero que  fator de 997799, que  {d}, porm no tem trs dgitos')
```

```python
# tentativa na fora bruta
palindromos = []
for i in range(900, 1000):
    for j in range(900, 1000):
        if (numero_palindromo(i*j)):
            palindromos.append([i*j, i, j])

palindromos = np.array(palindromos)
palindromos
```
