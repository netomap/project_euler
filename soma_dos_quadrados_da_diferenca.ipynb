- A soma dos quadrados dos números com a diferenaça com o quadrado das somas até 100

```python
soma_simples = 0
soma_quadrados = 0

for n in range(1, 100+1, 1):
    soma_simples += n
    soma_quadrados += (n*n)

quadrado_das_somas = soma_simples*soma_simples

print (f'{soma_simples=}')
print (f'{quadrado_das_somas=}')
print (f'{soma_quadrados=}')

diferenca = quadrado_das_somas-soma_quadrados
print (f'{diferenca=}')
```
