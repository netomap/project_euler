```python
def print_(m, end='\n'):
    if DEBUG: print (m, end=end)
```

```python
DEBUG = False

possibilidades = [[numeros[0]]]

for linha in range(1, 15):
    linha_atual = numeros[linha]
    ultimas_possibilidades = possibilidades[-1]

    print_ (f'{ultimas_possibilidades=}')
    print_ (f'{linha_atual=}')

    linha_possibilidades = []
    for catual, cadavalor in enumerate(linha_atual):
        print_ (f'{catual}:{cadavalor}', end='=>')
        temp = []
        for canterior, listaanterior in enumerate(ultimas_possibilidades):
            if ((catual - canterior) >=0 and (catual-canterior) <= 1):
                print_ (f'{listaanterior=}', end=', ')
                for vla in listaanterior:
                    temp.append(cadavalor+vla)
                
        print_ ('')
        
        linha_possibilidades.append(temp)
    
    print_ (f'{linha_possibilidades=}')
    possibilidades.append(linha_possibilidades)

    print_ (30*'*-')
```

```python
print ('Verificando o valor acumulado para cada linha')

for i, linha in enumerate(possibilidades):
    print (str(i).rjust(2), end='==')
    for k, v in enumerate(linha):
        print (f'{str(max(v)).rjust(4)}', end=' ')
    print ('')
```
