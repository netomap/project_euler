# Anlise

- Primeiro vamos avaliar, para cada palavra, quais as chances que cada uma tem para ter permutao com qualquer outra da lista

```python
def contagem_letras(palavra: str):
    """Funo que retorna tipo uma identidade da palavra. 
    Retorna a letra e sua ocorrncia, com as letras em ordem alfabtica
    """
    letras = sorted(set(list(palavra)))
    return ''.join([f'{l}{str(palavra).count(l)}' for l in letras])

for n in range(10):
    palavra = choice(words_list)
    print (n, palavra, contagem_letras(palavra), sep=' ==> ')
```

- vamos criar um dicionrio de cada palavra e sua "identidade", da funo contagem_letras

```python
identidades = dict()
for palavra in words_list:
    identidades[palavra] = contagem_letras(palavra)
```

```python
list(identidades.items())[:10]
```

- Agora vamos criar uma lista de possibilidades, comparando o resultado de cada palavra com as possibilidades das outras da word_list

```python
possibilidades = dict()
for palavra, identidade in identidades.items():
    possibilidades_da_palavra = []
    for palavra_aux, identidade_aux in identidades.items():
        if (palavra != palavra_aux and identidade == identidade_aux):
            possibilidades_da_palavra.append(palavra_aux)
    
    if len(possibilidades_da_palavra) > 0:
        possibilidades[palavra] = possibilidades_da_palavra
```

```python
list(possibilidades.items())[:25] # amostra dos 15 primeiros itens de possibilidades de anagramas "pares"
```

```python
def quadrado_perfeito(n):
    return sqrt(n) % 1 == 0

def obter_opcoes_permutacoes(palavra):
    """Retorna uma lista com um dicionrio, o nmero que representa a palavra e seu nmero.  
    Este nmero tambm precisa ser um quadrado perfeito.  
    """
    letras = sorted(list(set(palavra)))
    total_letras = len(letras)
    permutacoes = permutations(range(1, 10), total_letras)
    
    opcoes = []
    for p in permutacoes:
        aux = {}
        for k in range(len(letras)):
            aux[letras[k]] = p[k]
        
        numero = int(''.join([str(aux[letra]) for letra in palavra]))
        
        if (quadrado_perfeito(numero)):
            opcoes.append([aux, numero])
    
    return opcoes

obter_opcoes_permutacoes('dog')
```

```python
maior_valor = 0

for palavra, lista_palavras in possibilidades.items(): # para cada palavra e suas opes em todas as possibilidades
    palavra_gemea = lista_palavras[0] # pegamos a palavra gmea que  a primeira da lista
    lista_permutacoes = obter_opcoes_permutacoes(palavra) # pegamos a lista de permutaes de seus nmeros quadrados perfeitos
    
    for dicionario, numero_perfeito in lista_permutacoes: # para cada opo de dicionrio da lista de permutaes
        # recriamos o nmero da palavra gmea, a partir do dicionrio
        numero_palavra_gemea = int(''.join([str(dicionario[letra]) for letra in palavra_gemea])) 
        if quadrado_perfeito(numero_palavra_gemea): # se o nmero da palavra gmea tambm for um quadrado perfeito
            # ento esse nmero e seu "gmeo" atendem  condio
            
            print (f'{palavra}=>{numero_perfeito}, {dicionario=}, {palavra_gemea}=>{numero_palavra_gemea}')

            if (numero_palavra_gemea > maior_valor): # pegamos o maior valor
                maior_valor = numero_palavra_gemea
```
