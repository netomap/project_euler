- Qual é o primeiro índice de números de fibonacci que tem 1000 dígitos.

```python
numeros = [0, 1, 1]

n = 2
condicao = True
while condicao:
    numeros.append(numeros[-1]+numeros[-2])

    if (len(str(numeros[-1])) == 1000):
        condicao = False

    n += 1

print (len(numeros) - 1)
```
