```python
import numpy as np
from math import sqrt
from tqdm import tqdm
```

```python
def divisores(n: int)->list:
    return [d for d in range(1, n) if n % d == 0]

def soma_lista(l: list)->int:
    return sum(l)

print (soma_lista(divisores(284)), soma_lista(divisores(220)))
```

```python
possible_numbers = dict()
for k in tqdm(range(1, 10_000), ncols=100):
    possible_numbers[k] = soma_lista(divisores(k))
```

```python
amicable_numbers = []
for numero, soma_divisores in possible_numbers.items():
    if soma_divisores in possible_numbers.keys():
        if possible_numbers[soma_divisores] == numero and soma_divisores != numero:
            amicable_numbers.append(numero)

amicable_numbers
```

```python
sum(amicable_numbers)
```
