```python
from math import sqrt
from random import randint
```

```python
def listar_numeros_primos(n):
    """n  o total de inteiros primos que queremos na lista."""
    lista = [2]
    k = 3 # nmero inicial 3, j que 2 ns sabemos que  o primeiro primo
    while (len(lista) < n):
        if (k % 2 != 0): # se for mpar
            c = 0
            for d in range(1, int(sqrt(k)+1)):
                c += 1 if (k % d == 0) else 0
            
            if c == 1: lista.append(k)
        
        k += 1
    
    return lista

primos = listar_numeros_primos(2000)
print (f'primeiros primos: {primos[:20]}')
```

```python
def decompor_numero(n: int)->str:
    decomposicao = {p: 0 for p in primos}
    while n > 1:
        i = 0
        while n % primos[i] != 0:
            i+= 1
        decomposicao[primos[i]] += 1

        n //= primos[i]
    
    # d = {p: e for p, e in decomposicao.items() if e > 0}
    decomposicao_str = ' x '.join([f"{p}^{e}" for p, e in decomposicao.items() if e > 0])
    return decomposicao_str

decompor_numero(77)
```

```python
n = randint(2, 9999)
for k in range(n, n+25):
    print (k, decompor_numero(k), sep=' => ')
```
