```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
inicio = 1
fim = 100

numeros_com_8_divisores = []

for numero in range(inicio, fim+1, 1):
    total_divisores = 0
    for divisor in range(1, numero+1, 1):
        if numero % divisor == 0:
            total_divisores += 1
    
    if (total_divisores == 8):
        numeros_com_8_divisores.append(numero)

print (f'{len(numeros_com_8_divisores)=}')
```
