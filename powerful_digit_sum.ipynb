```python
import pandas as pd
```

```python
somas = []

for a in range(99+1):
    for b in range(99+1):
        resultado = pow(a, b)
        resultado_str = str(resultado)
        somas.append([a, b, sum([int(d) for d in str(pow(a, b))])])

df = pd.DataFrame(somas, columns=['a', 'b', 'soma'])

df.sort_values(by='soma', ascending=False).head()
```
