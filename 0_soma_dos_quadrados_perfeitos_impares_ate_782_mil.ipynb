- Dentre os primeiros 782 mil quadrados perfeitos, qual é a soma apenas dos ímpares.

```python
from math import sqrt
```

```python
def e_quadrado_perfeito(n: int)->bool:
    return sqrt(n) - int(sqrt(n)) == 0

def e_impar(n: int)->bool:
    return n % 2 != 0
```

```python
soma = 0
for n in range(1, 782000, 2):
    soma += n*n
soma
```
