```python
def classifica_numero(n: int)->str:
    n_str = str(n)
    tipo = ''
    for k in range(len(n_str)-1):
        if (n_str[k+1] > n_str[k]):
            tipo += 'c'
        if (n_str[k+1] < n_str[k]):
            tipo += 'd'
    
    return 'bouncy' if ('c' in tipo and 'd' in tipo) else tipo

classifica_numero(4321111)
```

```python
contador = 0
numero_final = 1000

for n in range(1, numero_final+1, 1):
    if (classifica_numero(n) == 'bouncy'):
        contador += 1

print (f'total de nmeros elsticos: {contador} at {numero_final}')
```

```python
bouncy_count = 0
n = 0
percentual = 0

while percentual < 0.99:
    n += 1

    if (classifica_numero(n) == 'bouncy'):
        bouncy_count += 1
    
    percentual = bouncy_count/n
    

print (f'{bouncy_count=}')
print (f'{n=}')
print (f'{percentual=}')
```
