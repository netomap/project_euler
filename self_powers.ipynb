- Fonte: [https://projecteuler.net/problem=48](https://projecteuler.net/problem=48)

- A principal solução está em usar a função pow(x, y, z), que retorna:
    - x**y quando usamos somente os dois primeiros parâmetros.
    - Retorna o resto de X**y / z. 

- A resposta para este problema é usar z=10*10, porque aí z teria 10 dígitos.
- E o resto de um número de 9 dígitos dividido por 10*10 é o próprio número.
- O resto de uma divisão de um número de 11 dígitos dividido por um 10**10, vão ser os dez últimos dígitos.

- Por exemplo:
    - 985 % 100 = 85 (porque são 9 inteiros de 100 e sobram 85)
    - 3456 % 1000 = 456 (porque são 3 inteiros de 1000 e sobram 456)

- Ou seja, podemos pegar os dez últimos dígitos de qualquer número usando essa função

```python
soma = 0

for k in range(1, 1_000 + 1):
    soma += k**k
```

```python
print (f'{str(soma)[-10:]}')
```
