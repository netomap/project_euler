```python
import numpy as np
from colorama import Fore
```

```python
def desenhar_rede(rede):
    
    print (rede + '\n')

    indices = 'abcdefghijklmnoprstuvwxyz'
    pontos = [str(p).strip() for p in str(rede).replace(")", "").replace("(", "").split(',')]

    valores = []
    for k in range(0, len(pontos), 2):
        ponto = pontos[k:k+2]
        vetor_indice = [indices.index(p) for p in ponto]
        valores.append(vetor_indice)

    tamanho_matriz = np.max(np.array(valores)) + 1
    
    matriz = np.zeros((tamanho_matriz, tamanho_matriz)).astype(int)
    
    for x, y in valores:
        matriz[x, y] = 1
        matriz[y, x] = 1

    for linha in matriz:
        for c in linha:
            print ((Fore.RED if c == 1 else Fore.WHITE) + str(c) + Fore.RESET, end=' ')
        print('')

rede = '(a, f), (f, e),  (c, b), (c, d)'
desenhar_rede(rede)
```
